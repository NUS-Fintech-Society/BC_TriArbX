```python
import json
from typing import *
```

```python
def readAndParseJSON(filepath: str) -> List[str]:
    data = None
    
    try:
        with open(filepath, "r") as f:
            data = json.load(f)
    except FileNotFoundError | TypeError:
        raise ValueError("Invalid filepath")
    else:
        names = [t["name"] for t in data["tokens"]]
        values = []
        
        for name in names:
            for name2 in names:
                if name != name2:
                     values.append([name, name2])
        
        return values
```

```python
readAndParseJSON("gemini-token-list.json")
```
